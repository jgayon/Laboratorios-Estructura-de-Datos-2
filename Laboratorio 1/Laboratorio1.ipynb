```python
# Clase ArbolAvl
class ArbolAvl:
    def __init__(self, root=None):
        self.root = root
    
    def insertar(self, key, data):
        num = int(data)
        self.root = self._insertar(self.root, num)

    def _insertar(self, key, data, node: Node, num) -> Node:
        if node is None:
            return Node(key, data)
        
        if num < node.numero:
            node.left = self._insertar(key, data, node.left, num)
        else:
            node.right = self._insertar(key, data, node.right, num)

        # Actualizar la altura del nodo actual
        node.height = 1 + max(self.altura(node.left), self.altura(node.right))

        # Obtener el factor de equilibrio del nodo para verificar el desequilibrio
        balance = self.balance(node)

        # Casos de desequilibrio y rotaciones
        #izquierda izquierda
        if balance > 1 and num < node.left.numero:
            return self.rotate_right(node)
        #derecha derecha
        if balance < -1 and num > node.right.numero:
            return self.rotate_left(node)
        #izquierda derecha
        if balance > 1 and num > node.left.numero:
            node.left = self.rotate_left(node.left)
            return self.rotate_right(node)
        #derecha izquierda
        if balance < -1 and num < node.right.numero:
            node.right = self.rotate_right(node.right)
            return self.rotate_left(node)
        
        return node

    def balance(self, node: Node) -> int:
        if node is None:
            return 0
        return self.altura(node.left) - self.altura(node.right)
    
    def altura(self, node: Node) -> int:
        if node is None:
            return 0
        return  node.height

    def rotate_left(self, a: Node) -> Node:
        b = a.right
        b.parent = a.parent
        a.right = b.left
        if a.right is not None:
            a.right.parent = a
        b.left = a
        a.parent = b
        if b.parent is not None:
            if b.parent.right == a:
                b.parent.right = b
            else:
                b.parent.left = b
        return b

    def rotate_right(self, a: Node) -> Node:
        b = a.left
        b.parent = a.parent
        a.left = b.right
        if a.left is not None:
            a.left.parent = a
        b.right = a
        a.parent = b
        if b.parent is not None:
            if b.parent.right == a:
                b.parent.right = b
            else:
                b.parent.left = b
        return b
    
    #Principal buscar por criterio
    def buscarcriterio(self, key = None, size = None):
        result = []
        if key is None and size is not None:
            result = self.search_by_size(self.root, size)
        elif key is not None and size is None:
            result = self.search_by_type(self.root, key)
        elif key is not None and size is not None:
            result = self.search_by_type_and_size(self.root, key, size)
        return result
    
    #Buscar por tamaño
    def search_by_size(self, node, size):
        result = []
        if node:
            if node.size == size:
                result.append(node)
            result.extend(self.search_by_size(node.left, size))
            result.extend(self.search_by_size(node.right, size))
        return result
    
    #Buscar por tipo
    def search_by_type(self, node, key):
        result = []
        if node:
            if node.key == key:
                result.append(node)
            result.extend(self.search_by_type(node.left, key))
            result.extend(self.search_by_type(node.right, key))
        return result
    
    #Buscar por tipo y tamaño
    def search_by_type_and_size(self, node, key, size):
        result = []
        if node:
            if node.key == key and node.size == size:
                result.append(node)
            result.extend(self.search_by_type_and_size(node.left, key, size))
            result.extend(self.search_by_type_and_size(node.right, key, size))
        return result
    
    def eliminar(self,key,data):
        pass
    def recorridoniveles(self):
        pass
    def buscar(self):
        pass

```
